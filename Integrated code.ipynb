# Loading and training data

```python
# Define the image directory
image_directory = 'C:\\Users\\user1\\Desktop\\DISSERTATION\\Models'

# Function to load and preprocess data for Model 1
def load_and_preprocess_data_model1(image_directory):
    # Add your code to load and preprocess data for Model 1
    SIZE = 224
    dataset = []  
    label = []  
    True_images = os.listdir(os.path.join(image_directory, 'Model 1\\True Images'))
    False_images = os.listdir(os.path.join(image_directory, 'Model 1\\False Images'))
    # Load parasitized images
    for image_name in True_images:
        if image_name.endswith('.png'):
            image = cv2.imread(os.path.join(image_directory, 'Model 1\\True Images', image_name))
            image = Image.fromarray(image, 'RGB')
            image = image.resize((SIZE, SIZE))
            dataset.append(np.array(image))
            label.append(1)
    # Load uninfected images
    for image_name in False_images:
        if image_name.endswith('.png'):
            image = cv2.imread(os.path.join(image_directory, 'Model 1\\False Images', image_name))
            image = Image.fromarray(image, 'RGB')
            image = image.resize((SIZE, SIZE))
            dataset.append(np.array(image))
            label.append(0)
    dataset = np.array(dataset)
    label = np.array(label)
    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(dataset, label, test_size=0.20, random_state=0)
    X_train = X_train / 255.
    X_test = X_test / 255.
    y_train = to_categorical(y_train)
    y_test = to_categorical(y_test)

    return X_train, X_test, y_train, y_test
```

```python
from tensorflow.keras.applications import MobileNetV2

# Function to load data based on selected model
def load_data(image_directory, selected_model):
    if selected_model == 'Model 1':
        X_train, X_test, y_train, y_test = load_and_preprocess_data_model1(image_directory)
    elif selected_model == 'Model 2':
        X_train, X_test, y_train, y_test = load_and_preprocess_data_model2(image_directory)
    elif selected_model == 'Model 3':
        X_train, X_test, y_train, y_test = load_and_preprocess_data_model3(image_directory)
    elif selected_model == 'Model 4':
        X_train, X_test, y_train, y_test = load_and_preprocess_data_model4(image_directory)
    else:
        messagebox.showerror("Error", "Invalid model selection")
        return None, None, None, None
    return X_train, X_test, y_train, y_test

# Function to get the model
def get_model(input_shape=(224, 224, 3)):
    # Add your code to create and compile the model
    vgg = vgg16.VGG16(weights='C:\\Users\\user1\\Desktop\\DISSERTATION\\vgg16_weights_tf_dim_ordering_tf_kernels_notop (1).h5', include_top=False, input_shape=input_shape)
    #vgg = vgg16.VGG16(weights= None, include_top=False, input_shape=input_shape)

    for layer in vgg.layers[:-5]:
        layer.trainable = False
    x = vgg.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(2, activation="softmax")(x)           
    model = Model(vgg.input, x)
   
    model.compile(loss="categorical_crossentropy", optimizer=SGD(learning_rate=0.0001, momentum=0.9), metrics=["accuracy"])
    return model
# Get and compile the model
model = get_model(input_shape=(224, 224, 3))
```

```python
# Function to train the selected model
def train_model():
    selected_model = selected_option.get()
    if selected_model == 'Model 1':
        train_model1()
    elif selected_model == 'Model 2':
        train_model2()
    elif selected_model == 'Model 3':
        train_model3()
    elif selected_model == 'Model 4':
        train_model4()
    else:
        messagebox.showerror("Error", "Invalid model selection")
        
# Update the train_model_click function to call the train_model function
def train_model_click():
    train_model()
```

```python
# Function to train and save Model 1
# Global variables
model_directory = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Trained Models"
trained_model = None  # Variable to store the trained model
def train_model1():
    global model, trained_model
    X_train, X_test, y_train, y_test = load_data(image_directory, 'Model 1')
    if X_train is not None:        
        update_status("Training Model 1...")        
        model = get_model(input_shape=(224, 224, 3))        
        history = model.fit(X_train, y_train, batch_size=16, epochs=100, verbose=1, validation_data=(X_test, y_test))
        
        loss = history.history['loss']
        val_loss = history.history['val_loss']
        epochs = range(1, len(loss) + 1)
        plt.plot(epochs, loss, 'y', label='Training loss')
        plt.plot(epochs, val_loss, 'r', label='Validation loss')
        plt.title('Training and validation loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()
    
        acc = history.history['accuracy']
        val_acc = history.history['val_accuracy']
        plt.plot(epochs, acc, 'y', label='Training acc')
        plt.plot(epochs, val_acc, 'r', label='Validation acc')
        plt.title('Training and validation accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.show()

        test_accuracy = None

        y_pred = np.argmax(model.predict(X_test), axis=1)
        cm = confusion_matrix(np.argmax(y_test, axis=1), y_pred)
        print("Confusion Matrix:")
        print(cm)
        sns.heatmap(cm, annot=True)
        plt.show()  # Display the heatmap  

       
        _, acc = model.evaluate(X_test, y_test)
        test_accuracy = round(acc * 100.0, 2)
        print("Accuracy = ", (acc * 100.0), "%")
        # Test on single image.
        n = 20  # Select the index of image to be loaded for testing
        img = X_test[n]
        plt.imshow(img)
        input_img = np.expand_dims(img, axis=0)  # Expand dims so the input is (num images, x, y, c)
        prediction = np.argmax(model.predict(input_img))
        actual_label = np.argmax(y_test[n]) 
        print("The prediction for this image is: ", np.argmax(model.predict(input_img)))
        print("The actual label for this image is: ", np.argmax(y_test[n]))
        plt.title(f'Prediction: {prediction}')
        plt.show()  # Display the image     
        update_status("Model 1 trained...")
        update_status(f'Test Accuracy: {test_accuracy}% | Predicted Label: {prediction} | Actual Label: {actual_label}' if actual_label is not None else "Model 3 trained successfully")


        # Set trained_model to the trained model
        trained_model = model
```

```python
def plot_heatmap(img):
    global model

    # Ensure image is in RGB color space
    if img.shape[-1] == 1:
        img = np.repeat(img, 3, axis=-1)

    pred = model.predict(np.expand_dims(img, axis=0))
    pred_class = np.argmax(pred)
    # Get weights for all classes from the prediction layer
    last_layer_weights = model.layers[-1].get_weights()[0]
    last_layer_weights_for_pred = last_layer_weights[:, pred_class]
    last_conv_model = Model(model.input, model.get_layer("block5_conv3").output)
    last_conv_output = last_conv_model.predict(np.expand_dims(img, axis=0))  # Predict on the image
    last_conv_output = np.squeeze(last_conv_output)  

    h = int(img.shape[0] / last_conv_output.shape[0])
    w = int(img.shape[1] / last_conv_output.shape[1])
    upsampled_last_conv_output = scipy.ndimage.zoom(last_conv_output, (h, w, 1), order=1)

    heat_map = np.dot(upsampled_last_conv_output.reshape((img.shape[0] * img.shape[1], 512)),
                      last_layer_weights_for_pred).reshape(img.shape[0], img.shape[1])

    heat_map[img[:, :, 0] == 0] = 0  # All dark pixels outside the object set to 0

    peak_coords = peak_local_max(heat_map, num_peaks=5, threshold_rel=0.5, min_distance=10)

    plt.imshow(img)
    plt.imshow(heat_map, cmap='jet', alpha=0.10)
    for i in range(0, peak_coords.shape[0]):
        y = peak_coords[i, 0]
        x = peak_coords[i, 1]
        plt.gca().add_patch(Rectangle((x - 15, y - 15), 50, 50, linewidth=1, edgecolor='r', facecolor='none'))

    # Save the heatmap image
    plt.savefig("heatmap.png")
    plt.close()

    # Open and resize the heatmap image
    heatmap_img = Image.open("heatmap.png")
    heatmap_img_resized = heatmap_img.resize((400, 400), Image.LANCZOS)

    #heatmap_img_resized = heatmap_img.resize((frame6.winfo_width(), frame6.winfo_height()), Image.LANCZOS)

    # Convert the heatmap image to a Tkinter-compatible format
    heatmap_image_tk = ImageTk.PhotoImage(heatmap_img_resized)

    # Display the heatmap image in the Tkinter label widget
    heatmap_label.config(image=heatmap_image_tk)
    heatmap_label.image = heatmap_image_tk

def test_model():
    global model
    X_train, X_test, y_train, y_test = load_data(image_directory, selected_option.get())
    if X_test is not None: 
        update_status("Testing model...")
        
        # Define test_accuracy variable
        test_accuracy = None
        
        _, acc = model.evaluate(X_test, y_test)
        test_accuracy = round(acc * 100.0, 2)
        print("Accuracy = ", (acc * 100.0), "%")
        
        # Test on single image.
        n = np.random.randint(0, X_test.shape[0])  # Select the index of image to be loaded for testing
        img = X_test[n]
        plt.imshow(img)
        input_img = np.expand_dims(img, axis=0)  # Expand dims so the input is (num images, x, y, c)
        prediction = np.argmax(model.predict(input_img))
        
        actual_label = np.argmax(y_test[n]) 
        
        
        update_status(f'Test Accuracy: {test_accuracy}% | Predicted Label: {prediction} | Actual Label: {actual_label}' if actual_label is not None else "Model tested successfully")
         
        print("The prediction for this image is: ", np.argmax(model.predict(input_img)))
        print("The actual label for this image is: ", np.argmax(y_test[n]))
        plt.title(f'Prediction: {prediction}')
        plt.show()  # Display the image 
        
        
        #Print confusion matrix
        y_pred = np.argmax(model.predict(X_test), axis=1)
        cm = confusion_matrix(np.argmax(y_test, axis=1), y_pred)
        print("Confusion Matrix:")
        print(cm)
        sns.heatmap(cm, annot=True)
        plt.show()  # Display the heatmap

        # Calculate metrics
        
        True_images_idx = np.where(y_pred == 1)[0]

        predicted_as_para=[]
        for i in True_images_idx:
            par_img = X_test[i]
            predicted_as_para.append(par_img)  
        predicted_as_para = np.array(predicted_as_para)
        
        update_status("Model tested successfully")
               
        import random
        im = random.randint(0, predicted_as_para.shape[0] - 1)
        heat_map = plot_heatmap(predicted_as_para[im])
        img = predicted_as_para[im]
        plt.imshow(predicted_as_para[im])
        plt.show()
```
