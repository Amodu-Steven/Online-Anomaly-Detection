```python
import tkinter as tk
from tkinter import ttk
import cv2
import os
import datetime
import numpy as np
import scipy  #Used to upsample our image
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def update_status(message):
    status_var.set(message)
```

```python
def update_camera():
    ret, frame = cap.read()
    if ret:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (380, 380))  # Resize the frame to fit the frame size
        img = Image.fromarray(frame)
        imgtk = ImageTk.PhotoImage(image=img)
        label_camera.imgtk = imgtk
        label_camera.config(image=imgtk)
    label_camera.after(10, update_camera)  # Update every 10 milliseconds
```

```python
def snap_picture():
    global still_image
    ret, frame = cap.read()
    if ret:
        still_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        still_image = cv2.resize(still_image, (280, 280))  # Resize for preview
        img = Image.fromarray(still_image)
        imgtk = ImageTk.PhotoImage(image=img)

        # Update the image in label_preview within frame2
        label_preview.imgtk = imgtk
        label_preview.config(image=imgtk)

        update_status("Still image captured")
```

# True Images

```python
def save_picture():
    global still_image
    if still_image is not None:
        selected_model = selected_option.get()  # Get the selected model
        if selected_model:
            timestamp = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
            filename = f"still_image_{timestamp}.png"
            if selected_model == "Model 1":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 1\\True Images"
            elif selected_model == "Model 2":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 2\\True Images"
            elif selected_model == "Model 3":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 3\\True Images"
            elif selected_model == "Model 4":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 4\\True Images"
                
            file_path = os.path.join(save_dir, filename)
            cv2.imwrite(file_path, cv2.cvtColor(still_image, cv2.COLOR_RGB2BGR))
            update_status(f"Still image saved as {file_path}")
        else:
            update_status("Please select a model")
    else:
        update_status("No still image captured")
```

```python
def save_picture_false():
    global still_image
    if still_image is not None:
        selected_model = selected_option.get()  # Get the selected model
        if selected_model:
            timestamp = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
            filename = f"still_image_{timestamp}.png"
            if selected_model == "Model 1":
                save_dir =  "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 1\\False Images"
            elif selected_model == "Model 2":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 2\\False Images"
            elif selected_model == "Model 3":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 3\\False Images"
            elif selected_model == "Model 4":
                save_dir = "C:\\Users\\user1\\Desktop\\DISSERTATION\\Models\\Model 4\\False Images"

            file_path = os.path.join(save_dir, filename)
            cv2.imwrite(file_path, cv2.cvtColor(still_image, cv2.COLOR_RGB2BGR))
            update_status(f"Still image saved as {file_path}")
        else:
            update_status("Please select a model")
    else:
        update_status("No still image captured")
```
